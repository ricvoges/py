```python
def recuperar_hashtags(texto):
    '#MachineLearning #Python #DataScience #BigData'
    palavras = texto.split()
    aux = []
    for plv in palavras:
        if plv.startswith('#'):
            aux.append(plv)
    return ' '.join(aux)
```

```python
print(recuperar_hashtags(df['text'].iloc[1]))
```

```python
df_aux['hashtags'] = df.text.apply(recuperar_hashtags)
```
